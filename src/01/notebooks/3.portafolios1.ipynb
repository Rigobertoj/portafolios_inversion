```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
prices = yf.download(["PLTR", "GLD", "CVX", "AAPL"], start="2021-01-01", end="2026-01-28")["Close"]
```

```python
prices
```

```python
returns = prices.pct_change().dropna()
returns
```

```python
# w = np.array([0.25, 0.25, 0.25, 0.25])
n = len(prices.columns)
n
```

```python
w = np.ones(n)
w
```

```python
w = w / n
w
```

```python
mu = returns.mean()
mu
```

```python
# Expected portfolio return
return_portfolio = mu @ w
return_portfolio
```

```python
return_portfolio * 252
```

```python
sigma = returns.std()
sigma
```

```python
sigma / mu
```

```python
cov_matrix = returns.cov()
cov_matrix
```

```python
var_portfolio = w.T @ cov_matrix @ w
var_portfolio
```

```python
vol_portfolio = np.sqrt(var_portfolio) * np.sqrt(252)
vol_portfolio
```

```python
# ESTO ESTA MAL
returns.std() @ w * np.sqrt(252)
```
