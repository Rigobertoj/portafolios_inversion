```python
df = pd.DataFrame(index=["Bear", "Neutral", "Bull"])
df["Probabilidades"] = [0.25, 0.30, 0.45]
df["Rendimiento"] = [-.10, .05, .20]
df
```

```python
E_r = (df["Probabilidades"] * df["Rendimiento"]).sum()
E_r*100
```

```python
df["Probabilidades"] @ df["Rendimiento"]
```

```python
E_var = ((df["Rendimiento"] - E_r) ** 2 * df["Probabilidades"]).sum() 

E_var
```

```python
E_vol = np.sqrt(E_var)

E_vol
```

```python
E_var, E_vol
```

```python
df2 = pd.DataFrame(index=["Dovish", "Neutra", "Hawkish"])
df2["P"] = [0.7, 0.2, 0.1]
df2["R"] = [0.02, 0.03, 0.055]

df2.head()
```

```python
Ev = df2["P"] @ df2["R"]
Ev
```

```python
Var = ((df2["R"] - Ev) ** 2) @ df2["P"]
Var * 100
```

```python
Vol = np.sqrt(Var)
Vol * 100
```
