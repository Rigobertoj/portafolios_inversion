```python
import pandas as pd
import numpy as np 

import yfinance as yf
```

```python
tickers = ["APPL", "MSFT", "LLY", "NVDA", "^GSPC"]

prices = yf.download(tickers,start="2020-01-01", end="2026-02-09")["Close"]
prices.head()
```

### Estrategia pasiva

```python
# Rendimiento, Riesgo de una estaregia pasiva
prices_pasiva=prices[['^GSPC']]
rets_daily=prices_pasiva.pct_change().dropna()

rend_pasiva=rets_daily.mean()*252
riesgo_pasiva=rets_daily.std()*np.sqrt(252)

rf=.0375

Sharpe_pasiva=(rend_pasiva-rf)/riesgo_pasiva
cv=riesgo_pasiva/rend_pasiva

rend_pasiva, riesgo_pasiva, Sharpe_pasiva, cv 
```

### Estrategia activa

```python
# Rendimiento y Riesgo de mi Estategia Activa 

w=np.array([.25,.25,.25,.25])

prices_activa=prices[['AAPL','MSFT','LLY','NVDA']]
rets_activa_daily=prices_activa.pct_change().dropna()

ind_means=rets_activa_daily.mean()
cov=rets_activa_daily.cov()

# INFO DEL PORTAFOLIO 
R_activa=sum(w*ind_means) *252
vol_activa=np.sqrt(w.T @ cov @ w) * np.sqrt(252)

Sharpe_activa=(R_activa-rf)/vol_activa
cv_activa=vol_activa/R_activa


Sharpe_activa, cv_activa
```
